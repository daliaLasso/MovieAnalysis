```python
# extract hours and minutes of run_time
df[['Hours', 'Minutes']] = df['run_time'].str.extract(r'(?:(\d+)h)? ?(?:(\d+)m)?')

# Replace NaN with 0 and convert to integers
df['Hours'] = df['Hours'].fillna(0).astype(int)
df['Minutes'] = df['Minutes'].fillna(0).astype(int)

# compute total minutes
df['TotalMinutes'] = df['Hours'] * 60 + df['Minutes']

# compute the average 
average_run_time = df['TotalMinutes'].mean()

print(f"Average time in minutes: {average_run_time}")
```
